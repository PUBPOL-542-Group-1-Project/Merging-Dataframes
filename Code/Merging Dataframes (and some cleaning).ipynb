Professor's Code to Transfrom DisciplineRate variable to a categorical variable.

```python
link='https://github.com/PUBPOL-542-Group-1-Project/Merging-Dataframes/raw/main/datafiles/merged_df.csv'

import pandas as pd

mergedData=pd.read_csv(link)
mergedData
```

```python
# column to recode
mergedData.DisciplineRate
```

```python
# values with "<"

mergedData.DisciplineRate[mergedData.DisciplineRate.str.contains("<")]
```

```python
# "Suppressed: N<10" will become missing value
import numpy as np

mergedData.replace(to_replace={'Suppressed: N<10':np.nan},inplace=True)
```

```python
# get the number in DisciplineRate
mergedData['number']=mergedData.DisciplineRate.str.extract(pat='(\d+\.*\d*)')
mergedData['number']
```

```python
# create a flag, so you remember where you had a '<'

mergedData['flag_<']=mergedData.DisciplineRate.str.contains("<")
mergedData['flag_<']
```

```python
#combine last to values

list(zip(mergedData.number, mergedData['flag_<']))
```

```python
# I will substract 0.1 to every "number" where the "flag" is True  

# notice the missings, the if-else, and the use of float()

[(float(number)-0.1) if flag==True else float(number) for number,flag  in zip(mergedData.number, mergedData['flag_<']) ]
```

```python
# I will use the same to create new column
mergedData['percent']=[(float(N)-0.1) if F else float(N) for N,F  in zip(mergedData.number, mergedData['flag_<']) ]
mergedData['percent']
```

```python
# I HAVE NUMBERS
# SEE basic stats
mergedData.percent.describe()
```

```python
# creating intervals after understanding the basic stats: (intervals that make sense)

mergedData['DisciplineInt']=pd.cut(mergedData.percent,
                                bins=[0,2,4,7,10,26], # cut points
                                labels=['below_2','2_to_4','4_to_7','7_10','above_10']) #labelling intervals
mergedData['DisciplineInt']
```

```python
#distribution of NEW variable
mergedData['DisciplineInt'].value_counts()
```

```python
# remember you created some variables
mergedData.iloc[:,-6:]
```

```python
# wanna drop the ones you do not need any more?
to_drop=['number','flag_<','percent']

mergedData.drop(columns=to_drop,inplace=True)
```

```python
mergedData.to_csv('merged_df2.csv',index=False)
```

Making final adjustments to the data

```python
#collecting data from github repository

import pandas as pd

location='https://github.com/PUBPOL-542-Group-1-Project/Merging-Dataframes/raw/main/datafiles/merged_df2.csv'
fin=pd.read_csv(location)
```

```python
#keep only variables we want for analysis
fin1=fin[['SchoolName','Graduate','Dropout','All Students','Male','White','English Language Learners','Low-Income','Expenditure','DisciplineInt']]
```

```python
fin2=fin1.copy()
```

```python
#find percentages
fin2['Male.p']=(fin2['Male']/fin2['All Students'])*100
fin2['White.p']=(fin2['White']/fin2['All Students'])*100
fin2['ELL.p']=(fin2['English Language Learners']/fin2['All Students'])*100
fin2['Low-Income.p']=(fin2['Low-Income']/fin2['All Students'])*100
```

```python
#delete rows with one blank 
fin2=fin2.dropna()
```

```python
fin2.info()
```

```python
#make disicpline rate ordinal factor
fin2.DisciplineInt.value_counts()
```

```python
#make copy
fin3=fin2.copy()

fin3['DisciplineInt']=pd.Categorical(fin3['DisciplineInt'],categories=["2_to_4", "below_2", "4_to_7","7_10","above_10"],ordered=True)
```
